```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sigmoid函数计算
def sigmoid(x):
    rst = 1.0/(1.0+np.exp(-x))
    return rst
```

```python
#计算交叉熵
def cal_cost(y_hat,y,m):
    cost = -1/m*(np.dot(y,np.log(y_hat).T)+
                 np.dot((1-y),np.log(1-y_hat).T))
    cost = np.squeeze(cost)
    return cost
```

```python
#计算dw,db
def cal_gradient(Y_hat,batch_y,batch_x,m):
    dz = (Y_hat-batch_y)/m
    dw = np.dot(dz,batch_x.T)
    db = np.sum(dz)
    return dw,db
```

```python
def train(X,Y,learning_rate,batch_size,max_iter):
    #将w初始化为1附近的随机数
    w = 0.1*np.random.randn(1,2)+1
    b = 0#b初始化为0
    n0,m = X.shape
    cost = []
    for epoch in range(max_iter):
        all_idxs = np.arange(m)
        np.random.shuffle(all_idxs)
        batch_num = round(m/batch_size)
        for i in range(batch_num):
            batch_idx = all_idxs[i*batch_size:(i+1)*batch_size]
            batch_x = X[:,batch_idx]
            batch_y = Y[:,batch_idx]
            Y_hat = sigmoid(np.dot(w,batch_x)+b)
            costi = cal_cost(Y_hat,batch_y,m)
            cost.append(costi)
            dw,db = cal_gradient(Y_hat,batch_y,batch_x,m)
            w -= learning_rate*dw
            b -= learning_rate*db
    return w,b,cost
```

```python
def gen_sample_data(data_size):
    n0,m = data_size
    x = np.random.random((n0,m))*4
    z = 2*x[0]+x[1]-5
    y = np.zeros((1,m),dtype = np.int8)
    #添加噪声
    x += np.random.random((n0,m))*np.random.randint(-1, 1)*0.2
    y[:,z>0] = 1
    return x,y
```

```python
def run():
    data_size = (2,100)
    learning_rate = 0.1
    batch_size = 16
    max_iter = 2000
    X_train,Y_train = gen_sample_data(data_size)
    #绘画训练数据
    color = np.array(['r' for _ in range(X_train.shape[1])])
    ytemp = np.squeeze(Y_train)
    color[ytemp>0] = 'g'
    #画出采样点,不同类别用颜色区分
    plt.scatter(X_train[0],X_train[1],c = color)
    plt.plot()
    #画出分类直线
    x_plt = np.linspace(0,4,40)
    y_plt = x_plt*(-2)+5
    plt.plot(x_plt,y_plt)
    plt.axis([0,4,0,4])
    w,b,cost = train(X_train,Y_train,learning_rate,batch_size,max_iter)
    y_new = (x_plt*w[0,0]+b)/(-w[0,1])
    plt.plot(x_plt,y_new,c='k')
    plt.show()
    #画出cost函数
    x_lable = np.arange(len(cost))
    plt.plot(x_lable,cost)
    plt.show()
```

```python
if __name__ == '__main__':
    run()
```

- 当采用mini-batch方式进行训练是,cost函数会有较多的波动,当batch_size等于训练样本数量时(不采用mini-batch方式)时,cost函数是平滑的.
- 但是相同的迭代次数,mini-batch方式具有更快的学习速度,cost函数能够达到更低的值.
