### 第二周编程编程作业

关于图像中值滤波的实现.主要实现了两种方法,一种原始的给予排序的方法medianBlur(),排序算法比较了Python自带的sort()方法和自己实现的快排算法.另一种是TS.Huang等提出的改进的基于灰度直方图的方法medianBlur_hist().对两种方法的性能进行了比较,并对结果进行了分析.

```python
import cv2
import numpy as np
import time
```

```python
def fastsort(x):
    xlen = len(x)
    if xlen == 1:
        return x
    else:
        midid = int(xlen/2)
        midx = x[midid]
        left = []
        mid = []
        right = []
        for i in range(xlen):
            if x[i]<midx:
                left.append(x[i])
            elif x[i] == midx:
                mid.append(x[i])
            else:
                right.append(x[i])
        if len(left) > 0 and len(right)>0:
            sortl = fastsort(left)
            sortr = fastsort(right)
            sortl.extend(mid)
            sortl.extend(sortr)
            return sortl
        elif len(left) == 0 and len(right) == 0:
            return mid
        elif len(left) > 0:
            sortl = fastsort(left)
            sortl.extend(mid)
            return sortl
        else:
            sortr = fastsort(right)
            mid.extend(sortr)
            return mid
```

```python
def getmidpix(win):#给出滤波窗口像素值,给出中值像素
    h,w = win.shape
    line_win = win.reshape(h*w)
    line_win = np.squeeze(line_win)
    #运用快排算法进行排序
#     sorted_win = fastsort(line_win)
#     return sorted_win[int(h*w/2)]
    line_win.sort()
    return line_win[int(h*w/2)]
    
```

```python
def padding(img,padding_way,col,row):
    h,w = img.shape
    rst_img = np.zeros((h+2*col,w+2*row),np.int)
    rst_img[col:-col,row:-row] = img
    if padding_way == 'ZERO':
        return rst_img
    else:
        #上&下边带
        for i in range(col):
            rst_img[i] = rst_img[col]
            rst_img[-1-i] = rst_img[-1-col]
        for j in range(row):
            rst_img[:,j] = rst_img[:,row]
            rst_img[:,-1-j] = rst_img[:,-1-row]
    return rst_img
```

```python
#运用初级的二维窗口进行中值滤波
def medianBlur(img, kernel, padding_way):
    h,w = img.shape
    m,n = kernel.shape
    mhalf = int(m/2)
    nhalf = int(n/2)
    rst_img = img.copy()
    if m%2 == 0 or n%2 == 0:
        return 'bad kernel'
    #padding image,默认padding size为滤波窗口的一半
    img_padding = padding(img,padding_way,mhalf,nhalf)
    #median filter
    for i in range(h):
        for j in range(w):
            rst_img[i][j] = getmidpix(img_padding[i:i+m,j:j+n])
    return rst_img
```

```python
#运用统计直方图的方式,进行图像中值滤波
def medianBlur_hist(img,kernel, padding_way):
    h,w = img.shape
    m,n = kernel.shape
    mhalf = int(m/2)
    nhalf = int(n/2)
    midnum = int(m*n/2)+1
    rst_img = img.copy()
    if m%2 == 0 or n%2 == 0:
        return 'bad kernel'
    #padding image,默认padding size为滤波窗口的一半
    img_padding = padding(img,padding_way,mhalf,nhalf)    
    #median filter
    hist = np.zeros((h,256))#每行第一个window的直方图,用于向右侧移动时进行递推下一个直方图
    hm = [0 for _ in range(h)]#存放每行第一个window的中值
    hsum = [0 for _ in range(h)]#存放中值左侧直方图累加值
    j = 0
    for i in range(h):
        if i == 0:
            win = img_padding[i:i+m,j:j+n]
            #统计直方图
            for t in range(m):
                for k in range(n):
                    hist[i][win[t][k]] += 1
            #计算中位数
            lsum = 0
            for l in range(256):
                lsum += hist[i][l]
                if lsum >= midnum:
                    break
            hm[i] = l
            hsum[i] = lsum
            rst_img[i,j] = hm[i]
        else:
            #更新直方图
            hist[i] = hist[i-1]
            mid = hm[i-1]
            sm = hsum[i-1]
            t = i-1
            for k in range(n):
                hist[i][img_padding[t][k]] -= 1
                if img_padding[t][k] <= mid:
                    sm -= 1
            t = i+m-1
            for k in range(n):
                hist[i][img_padding[t][k]] += 1
                if img_padding[t][k] <= mid:
                    sm += 1
            #计算中值
            if sm>=midnum:
                for l in range(mid,-1,-1):
                    sm -= hist[i][l]
                    if sm < midnum:
                        hm[i] = l
                        hsum[i] = sm+hist[i][l]
                        break
            else:
                for l in range(mid+1,256):
                    sm += hist[i][l]
                    if sm >= midnum:
                        hm[i] = l
                        hsum[i] = sm
                        break
            rst_img[i,j] = hm[i]
    for i in range(h):
        histrow = hist[i]
        mid = hm[i]
        sm = hsum[i]
        for j in range(1,w):
            #更新直方图
            k = j-1
            for t in range(m):
                histrow[img_padding[t+i][k]] -= 1
                if img_padding[t+i][k] <= mid:
                    sm -= 1
            k = j+n-1
            for t in range(m):
                histrow[img_padding[t+i][k]] += 1
                if img_padding[t+i][k] <= mid:
                    sm += 1                
            #计算中值
            if sm>=midnum:
                for l in range(mid,-1,-1):
                    sm -= histrow[l]
                    if sm < midnum:
                        mid = l
                        sm = sm+histrow[l]
                        break
            else:
                for l in range(mid+1,256):
                    sm += histrow[l]
                    if sm >= midnum:
                        mid = l
                        break
            rst_img[i,j] = mid
    return rst_img
```

```python
#对输入图像进行测试
img_gray = cv2.imread('lena.tif', 0)
#img_gray = cv2.resize(img_gray,(8,8))
kernel = np.ones((5,5),np.int)
padding_way = 'REPLICA'
t1 = time.time()
timg1 = medianBlur(img_gray, kernel, padding_way)
t2 = time.time()
print('orig:',t2-t1)
t1 = time.time()
timg2 = medianBlur_hist(img_gray, kernel, padding_way)
t2 = time.time()
print('hist:',t2-t1)
cv2.imshow('lena', img_gray)
cv2.imshow('filter_orig',timg1)
cv2.imshow('filter_hist',timg2)
key = cv2.waitKey()
if key == 27:
    cv2.destroyAllWindows()
```

### 结论

每个滑窗排序求得中值的方法,比TS.Huang等提出的利用灰度直方图改进中值滤波方法运行时间更快.分析这种反常原因主要有以下几点:

1. 原始方法中排序使用python自带的sort()函数,该函数经过高度优化,运行效率较高,如换成自己写的快排算法,则运行时间比改进中值滤波算法要慢很多.
2. 改进的利用直方图的方法,其中很多步骤可以并行处理,本程序中利用此优势
3. 改进的灰度直方图算法在实现过程中,灰度直方图实现的数据结构可能并不完美,程序中因为灰度范围为[0,255],所以直接定义了一个长度为256的列表,这样导致每个滑窗的灰度直方图有很多为0的部分,白白浪费了较多的计算,但是暂未发现有效的解决办法.
